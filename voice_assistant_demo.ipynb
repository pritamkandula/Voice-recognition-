```python
import pyttsx3 as p
import speech_recognition as sr
import randfacts
import datetime
import requests
from selenium import webdriver    
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By


engine = p.init()
rate = engine.getProperty('rate')
engine.setProperty('rate',180)
voices = engine.getProperty('voices')
engine.setProperty('voice',voices[0].id)

#wikipedia
class infow():
    def __init__(self):
        self.driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
        
    def get_info(self,query):
        self.query = query
        self.driver.get(url='https://www.wikipedia.org')
        search = self.driver.find_element(By.XPATH, '//*[@id="searchInput"]' )
        search.send_keys(query)
        search.send_keys(Keys.RETURN)


#google
class surf():
    def __init__(self):
        self.driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
        
    def get_info(self,query):
        self.query = query
        self.driver.get(url='https://www.google.com/')
        search = self.driver.find_element(By.XPATH, '//*[@name="q"]' )
        search.send_keys(query)
        search.send_keys(Keys.RETURN)

#youtube
class music():
    def __init__(self):
        self.driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
        
    def play(self,query):
        self.query = query
        self.driver.get(url='https://www.youtube.com/results?search_query=' + query)
        video = self.driver.find_element(By.XPATH, '//*[@id="title-wrapper"]' )
        video.click()

#jokes
def joke():
    url = 'https://v2.jokeapi.dev/joke/Any?type=twopart'
    json_data = requests.get(url).json()

    jk = ["",""]
    jk[0] = json_data['setup']
    jk[1] = json_data['delivery']
    return jk

#news
def news():
    url = ('https://newsapi.org/v2/top-headlines?'
       'country=us&'
       'apiKey=7fdf90c0c7cc4ab997e0cf299fb7565f')
    response = requests.get(url).json()
    ar = []
    for i in range(5):
        ar.append('Number' + str(i+1) + ", " + response['articles'][i]['title'] + ".")
    return ar

def speak(text):
    engine.say(text)
    engine.runAndWait()
    
def wishme():
    hour = int(datetime.datetime.now().hour)
    if hour>0 and hour<12:
        return('Morning bro')
    elif hour>=12 and hour<16:
        return('Good Afternoon')
    else:
        return('Good Evening')

today_date = datetime.datetime.now()

r = sr.Recognizer()  

speak(' Hey Pritham, '+wishme()+', I am your voice assistant Haazaard, how are you?')

with sr.Microphone() as source:
    r.energy_threshold = 10000   
    r.adjust_for_ambient_noise(source,1.2)
    print('Listening')
    audio = r.listen(source)
    text = r.recognize_google(audio)
    print(str(text))
if 'what' and 'about' and 'you' in text:
    speak("I'm doing good")
speak("What can I do for you Pritham? ")

with sr.Microphone() as source:
    r.energy_threshold = 10000   
    r.adjust_for_ambient_noise(source,1.2)
    print('Listening......')
    audio = r.listen(source)
    text2 = r.recognize_google(audio)
    print(text2)

if 'information' in text2:
    speak('you need information related to which topic?')
    
    with sr.Microphone() as source:
        r.energy_threshold = 10000   
        r.adjust_for_ambient_noise(source,1.2)
        print('Listening......')
        audio = r.listen(source)
        infor = r.recognize_google(audio)
    speak(' ok! let me search {} in wikipedia '.format(infor)) 
    print(' ok! let me search {} in wikipedia '.format(infor)) 
    
    assist = infow()
    assist.get_info(infor)
    
elif ('play' or 'video' or 'song' or 'music') in text2:
    speak('what do you want me to play?')
     
    with sr.Microphone() as source:
        r.energy_threshold = 10000   
        r.adjust_for_ambient_noise(source,1.2)
        print('Listening......')
        audio = r.listen(source)
        vid = r.recognize_google(audio)
    speak(' just a min!.... And there you go ')
    print('Playing {} on youtube'.format(vid))
    
    assist = music()
    assist.play(vid)
    
elif ('news') in text2:
    speak('Sure Pritham!, let me read it for you')
    arr = news()
    for i in range(len(arr)):
        print(arr[i])
        speak(arr[i])
        
        arr = news()
        print(arr)
        
elif ('fact' or 'facts') in text2:
    x = randfacts.getFact()
    print(x)
    speak('Did you know that, ' + x)
    
elif ('joke' or 'jokes') in text2:
    speak("Sure brooo!, don't complain if your stomach aches! ")
    jk = joke()
    print(jk[0])
    speak(jk[0])
    print(jk[1])
    speak(jk[1])
    
elif ('date' or 'day' or 'month' or 'year') in text2:
    today_date = datetime.datetime.now()
    print('today is ' + today_date.strftime('%d') + ' of ' + today_date.strftime('%B') + ", And it's the year " + today_date.strftime('%Y'))
    speak('today is ' + today_date.strftime('%d') + ' of ' + today_date.strftime('%B') + ", And it's the year " + today_date.strftime('%Y')) 


    
else:
    speak('Is there anything that i need to search on Google?')
    
    with sr.Microphone() as source:
        r.energy_threshold = 10000   
        r.adjust_for_ambient_noise(source,1.2)
        print('Listening......')
        audio = r.listen(source)
        goog = r.recognize_google(audio)
    speak(' ok! let me show you the searches for {} in Google '.format(goog)) 
    print(' ok! let me show you the searches for {} in Google '.format(goog)) 
    
    dekh = surf()
    dekh.get_info(goog)
```
